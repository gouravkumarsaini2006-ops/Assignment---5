6. Create a DataFrame with some missing values intentionally placed in the 
Marks and City columns. Then:  
• Identify missing values using isnull() 
• Fill missing marks with the column mean using fillna() 
• Drop rows where City is missing using dropna() 

```python

# Ans 6.
student_data = {
    "Name": ["Aman", "Riya", "Karan", "Neha", "Rahul", "Pooja"],
    "Marks": [85, np.nan, 78, 92, np.nan, 88],
    "City": ["Delhi", "Mumbai", np.nan, "Pune", "Jaipur", np.nan]
}

df = pd.DataFrame(student_data)

print(df)

print(df.isnull())

average_marks = df["Marks"].mean()

df["Marks"] = df["Marks"].fillna(average_marks)

print(df)

df = df.dropna(subset=["City"])

print(df)
```

7. Read a CSV file (you may create it manually) containing product sales 
data with columns: Product, Category, Sales, Region. Perform:  
• Group by Category and calculate total Sales using groupby() and agg() 
• Find the top 3 products by sales using sort_values() and head()

```python
# Ans 7.
sales = pd.read_csv(""sales.csv".csv")

category_sales = sales.groupby("Category").agg({"Sales": "sum"})

print(category_sales)

top_products = sales.sort_values("Age", ascending=False).head(3)

print(top_products)
```

8. Merge two DataFrames:  
• DataFrame 1: StudentID, Name, Department 
• DataFrame 2: StudentID, Marks, Grade 
• Merge on StudentID and display students who scored Grade 'A'

```python
# Ans 8.

df1 = pd.DataFrame({
    "StudentID": [101, 102, 103, 104],
    "Name": ["Aman", "Riya", "Karan", "Neha"],
    "Department": ["CSE", "IT", "ECE", "CSE"]
})

df2 = pd.DataFrame({
    "StudentID": [101, 102, 103, 104],
    "Marks": [92, 75, 88, 95],
    "Grade": ["A", "B", "A", "A"]
})

merged_df = pd.merge(df1, df2, on="StudentID")

print(merged_df)

a_grade_students = merged_df[merged_df["Grade"] == "A"]

print(a_grade_students)
```

10)   Using Pandas built-in plotting, load or create a dataset with at least 
4 numeric columns. Create:  
• A histogram for marks distribution 
• A box plot to visualize spread and outliers 
• A pie chart showing category-wise percentage

```python
# Ans 10
data = {
    "Math": [85, 90, 78, 92, 88, 81, 76, 95],
    "Science": [80, 88, 75, 90, 84, 79, 72, 94],
    "English": [82, 87, 74, 89, 83, 77, 70, 91],
    "Computer": [95, 97, 90, 99, 94, 88, 85, 100]
}

df = pd.DataFrame(data)

df["Math"].plot(kind="hist", bins=5)
plt.title("Marks Distribution")
plt.show()

df.plot(kind="box")
plt.title("Box Plot")
plt.show()

categories = pd.Series(
    [40, 30, 20, 10],
    index=["Electronics", "Clothing", "Footwear", "Accessories"]
)

categories.plot(kind="pie", autopct="%1.1f%%")
plt.title("Category-wise Percentage")
plt.ylabel("")
plt.show()
```

11)Using Seaborn, create a dataset of at least 50 records with columns 
Age, Salary, Department, Experience. Plot:  
• A scatter plot of Age vs Salary with hue as Department 
• A heatmap showing correlation between numeric columns 
• A bar plot showing average Salary per Department 

```python
# Ans 11
np.random.seed(1)

df = pd.DataFrame({
    "Age": np.random.randint(22, 50, 50),
    "Salary": np.random.randint(25000, 100000, 50),
    "Department": np.random.choice(["IT", "HR", "Finance", "Marketing"], 50),
    "Experience": np.random.randint(1, 20, 50)
})

sns.scatterplot(data=df, x="Age", y="Salary", hue="Department")
plt.title("Age vs Salary")
plt.show()
```

```python
corr_matrix = df[["Age", "Salary", "Experience"]].corr()

sns.heatmap(corr_matrix, annot=True)

plt.title("Correlation Heatmap")
plt.show()
```

12) Create a subplot layout of 2 rows × 2 columns using Matplotlib 
containing:  
• Line plot 
• Scatter plot 
• Bar chart 
• Histogram 
All using the same dataset. Add a main title using suptitle() 

```python
# Ans 12.
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales": [100, 150, 200, 180, 250],
    "Profit": [20, 35, 40, 38, 50]
}

df = pd.DataFrame(data)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(df["Month"], df["Sales"])
axes[0, 0].set_title("Line Plot")

axes[0, 1].scatter(df["Sales"], df["Profit"])
axes[0, 1].set_title("Scatter Plot")

axes[1, 0].bar(df["Month"], df["Sales"])
axes[1, 0].set_title("Bar Chart")

axes[1, 1].hist(df["Sales"])
axes[1, 1].set_title("Histogram")

plt.suptitle("Different Types of Plots")

plt.tight_layout()

plt.show()
```
